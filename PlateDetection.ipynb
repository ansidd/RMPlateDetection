<h1>Preparing Dataset</h1>
To do transfer learning on yolov5 model the datset must be in darknet format. Taking the images and annotations in xml files from here : "https://drive.google.com/drive/folders/1w4KQpC82ST2jVguzBdu4wQb6ceWD1dZf?usp=sharing" forming a dataset in the darknet format

```python
import sys
import os
import xml.etree.ElementTree as ET
import xmltodict


files = os.listdir('./labels_backup/')

yolo_inp_width = 640
yolo_inp_height = 640

for file in files:
    if '.xml' in file:
        with open('./labels_backup/{}'.format(file)) as f:
            xml_file = xmltodict.parse(f.read())
            
        image_name = file[:-4]
        objects = []
        print(xml_file)
        im_width, im_height = int(xml_file['annotation']['size']['width']), int(xml_file['annotation']['size']['height'])
        anns = xml_file['annotation']['object']
        if type(anns) is dict:
            anns = [anns]
        for obj in anns:
            xmin, ymin, xmax, ymax = int(obj['bndbox']['xmin']), int(obj['bndbox']['ymin']), int(obj['bndbox']['xmax']), int(obj['bndbox']['ymax'])
            xcenter = ((xmin+xmax)//2 * (yolo_inp_width)/im_width)/yolo_inp_width
            ycenter = ((ymin+ymax)//2 * (yolo_inp_height)/im_height)/yolo_inp_height
            bb_width = ((xmax - xmin) * (yolo_inp_width)/im_width)/yolo_inp_width
            bb_height = ((ymax - ymin) * (yolo_inp_height)/im_height)/yolo_inp_height
            
            objects.append('{} {} {} {} {}'.format(0,xcenter, ycenter, bb_width, bb_height))
        print(objects)
        with open('./data yolo format/labels/{}.txt'.format(image_name),'w') as f:
            f.write('\n'.join(objects))
            
        #os.rename('./data yolo format/images/{}'.format(file), './labels_backup/{}'.format(file))


# for folder in os.listdir('./data/'):
#     try:
#         for file in os.listdir('./data/{}/'.format(folder)):
#             source_path = "./data/{}/{}".format(folder,file)
#             file_name = file[:-4]
#             new_file_name = file_name+'_'+folder+file[-4:]
#             dest_path = "./data yolo format/images/{}".format(new_file_name)
#             os.rename(source_path, dest_path)
#     except Exception as e:
#         continue

data_list = os.listdir('./data yolo format/images/')
import random
random.shuffle(data_list)

l = len(data_list)
train_indices = (0,int(l*0.6))
test_indices = (train_indices[1], int(l*0.8))
val_indices = (test_indices[1], l)
train = data_list[train_indices[0]:train_indices[1]]
test = data_list[test_indices[0]:test_indices[1]]
val = data_list[val_indices[0]:val_indices[1]]

for i in train:
    try:
        file_name = i[:-4]
        #os.rename('./data yolo format/images/{}'.format(i),'./data yolo format/images/train/{}'.format(i) )
        os.rename('./data yolo format/labels/{}'.format(file_name+'.txt'),'./data yolo format/labels/train/{}'.format(file_name+'.txt') )
    except Exception as e:
        print(file_name)
        continue

for i in test:
    file_name = i[:-4]
    #os.rename('./data yolo format/images/{}'.format(i),'./data yolo format/images/test/{}'.format(i) )
    os.rename('./data yolo format/labels/{}'.format(file_name+'.txt'),'./data yolo format/labels/test/{}'.format(file_name+'.txt') )

for i in val:
    try:
        file_name = i[:-4]
        #os.rename('./data yolo format/images/{}'.format(i),'./data yolo format/images/val/{}'.format(i) )
        os.rename('./data yolo format/labels/{}'.format(file_name+'.txt'),'./data yolo format/labels/val/{}'.format(file_name+'.txt') )
    except Exception as e:
        print(file_name)
        continue

for t in ['train','test','val']:
    print(len(os.listdir("./data yolo format/images/{}".format(t))))
    print(len(os.listdir("./data yolo format/labels/{}".format(t))))

for t in ['train','test','val']:
    images = os.listdir("./data yolo format/images/{}".format(t))
    labels = os.listdir("./data yolo format/labels/{}".format(t))
    for i in images:
        file_name = i[:-4]
        if file_name+'.txt' not in labels:
            print(file_name)
            #os.remove("./data yolo format/images/{}/{}".format(t,i))

for t in ['train','test','val']:
    images = os.listdir("./data yolo format/images/{}".format(t))
    labels = os.listdir("./data yolo format/labels/{}".format(t))
    for i in labels:
        file_name = i[:-4]
        if file_name+'.jpg' not in images:
            print(file_name)
            #os.remove("./data yolo format/labels/{}/{}".format(t,i))
```

<h1>Making inference on an image for testing</h1>

```python
import torch

model = torch.hub.load('ultralytics/yolov5', 'custom', './best.pt')

import cv2

img = cv2.imread('./data yolo format/images/test/100_batch3.jpg')

import matplotlib.pyplot as plt
%matplotlib inline

plt.imshow(img)

res = model(img)

img = cv2.imread('./data yolo format/images/test/100_batch3.jpg')
for i in res.pred[0]:
    i = i.detach().cpu().numpy()
    x1, y1, x2, y2 = i[:4]

    img = cv2.rectangle(img, (int(x1),int(y1)), (int(x2),int(y2)), (255,0,0), 3)
plt.imshow(img)
```
